```python
grammar = r"""
Program → StmtList
StmtList → Stmt StmtList'
StmtList' → ; Stmt StmtList' | ε
Stmt → id := Expr
Expr → Term Expr'
Expr' → + Term Expr' | ε
Term → id | number
"""
```

```python
"""
Conjunto First

FIRST(Program)   = { id }
FIRST(StmtList)  = { id }
FIRST(StmtList') = { ; , ε }
FIRST(Stmt)      = { id }
FIRST(Expr)      = { id, number }
FIRST(Expr')     = { + , ε }
FIRST(Term)      = { id, number }
"""
```
